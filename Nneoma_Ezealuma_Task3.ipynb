## No.9

**Find the sales made in each month. Any trend noticeable?**


```python
monthly_sales = Compiled_data.groupby('Month_in_words')['Revenue'].sum().sort_values() # Sorted by sales amount
# Or, to sort by month order (if using Month_Name and want chronological order)

month_order = ['January']

monthly_sales_ordered = Compiled_data.groupby('Month_in_words')['Revenue'].sum().loc[month_order]
```

```python

monthly_sales_ordered.plot(kind='line', figsize=(10, 6))

plt.title('Monthly Sales Trend')

plt.xlabel('Month_in_words')

plt.ylabel('Revenue')

plt.grid(True)

plt.show()
```

```python
# Calculate month-over-month percentage change (requires a time-indexed Series)

monthly_sales_series = monthly_sales_ordered.reset_index().set_index('Month_in_words')['Revenue']

monthly_sales_series.index = pd.CategoricalIndex(monthly_sales_series.index, categories=month_order, ordered=True)

monthly_sales_series = monthly_sales_series.sort_index()


month_over_month_change = monthly_sales_series.pct_change() * 100

print("\nMonth-over-month percentage change:")

print(month_over_month_change)
```
